```python
import matplotlib.pyplot as plt


class MyPerceptron:
    def __init__(self):
        self.w = None
        self.b = 0
        self.l_rate = 1

    def fit(self, X_train, y_train):
        # ç”¨æ ·æœ¬ç‚¹çš„ç‰¹å¾æ•°æ›´æ–°åˆå§‹wï¼Œå¦‚x1=(3,3)Tï¼Œæœ‰ä¸¤ä¸ªç‰¹å¾ï¼Œåˆ™self.w=[0,0]
        self.w = np.zeros(X_train.shape[1])
        i = 0
        while i < X_train.shape[0]:
            X = X_train[i]
            y = y_train[i]
            # å¦‚æœy*(wx+b)â‰¤0 è¯´æ˜æ˜¯è¯¯åˆ¤ç‚¹ï¼Œæ›´æ–°w,b
            if y * (np.dot(self.w, X) + self.b) <= 0:
                self.w = self.w + self.l_rate * np.dot(y, X)
                self.b = self.b + self.l_rate * y
                i = 0  # å¦‚æœæ˜¯è¯¯åˆ¤ç‚¹,ä»å¤´è¿›è¡Œæ£€æµ‹
            else:
                i += 1


def draw(X, w, b):
    # äº§ç”Ÿåˆ†ç¦»è¶…å¹³é¢ä¸Šçš„ä¸¤ç‚¹
    X_new = np.array([[0], [6]])
#     åŸè§†é¢‘é”™è¯¯å¤„å·²æ›´æ–°
    y_predict = -(b + w[0] * X_new) / w[1] #w0x+w1y+b=0 åˆ™y=-b/w1-w0*x/w1
    print(y_predict)
    # ç»˜åˆ¶è®­ç»ƒæ•°æ®é›†çš„æ•£ç‚¹å›¾
    plt.plot(X[:2, 0], X[:2, 1], "g*", label="1")
#     åŸè§†é¢‘é”™è¯¯å¤„ç†æ›´æ–°
    plt.plot(X[2:, 0], X[2:, 1], "rx", label="-1")
    # ç»˜åˆ¶åˆ†ç¦»è¶…å¹³é¢
    plt.plot(X_new, y_predict, "b-")
    # è®¾ç½®ä¸¤åæ ‡è½´èµ·æ­¢å€¼
    plt.axis([0, 6, 0, 6])
    # è®¾ç½®åæ ‡è½´æ ‡ç­¾
    plt.xlabel('x1')
    plt.ylabel('x2')
    # æ˜¾ç¤ºå›¾ä¾‹
    plt.legend()
    # æ˜¾ç¤ºå›¾åƒ
    plt.show()


def main():
    # æ„é€ è®­ç»ƒæ•°æ®é›†
    X_train = np.array([[3, 3], [4, 3], [1, 1]])
    y_train = np.array([1, 1, -1])
    # æ„å»ºæ„ŸçŸ¥æœºå¯¹è±¡ï¼Œå¯¹æ•°æ®é›†ç»§ç»­è®­ç»ƒ
    perceptron = MyPerceptron()
    perceptron.fit(X_train, y_train)
    print(perceptron.w)
    print(perceptron.b)
    # ç»“æœå›¾åƒç»˜åˆ¶
    draw(X_train, perceptron.w, perceptron.b)


if __name__ == "__main__":
    main()
```

```python
def loadfile(filename):
    dataArr = []
    labelArr = []
    file = open(filename,"r")
    for line in file.readlines():
        currLine = line.strip().split(",")
        if int(currLine[0]) >= 5:#äºŒåˆ†ç±»é—®é¢˜
            labelArr.append(1)
        else:
            labelArr.append(-1)
        dataArr.append([int(num)/255 for num in currLine[1:]])
    return dataArr,labelArr

def perceptron(dataArr,labelArr,iter=50):
    dataMat = np.mat(dataArr)
    labelMat = np.mat(labelArr).T
    m,n = np.shape(dataMat)
    w = np.zeros((1,n))
    b = 0
    h = 0.0001
    
    for k in range(iter):
        for i in range(m):
            xi = dataMat[i]#returns vector
            yi = labelMat[i]
            if -1*yi*(w*xi.T+b) >= 0:
                w+=h*yi*xi
                b+=h*yi
        print('Round %d:%d training' % (k, iter))
    print("w is : {}".format(w))
    print("b is : {}".format(b))
    return w,b
def modelTest(dataArr,labelArr,w,b):
    dataMat = np.mat(dataArr)
    labelMat = np.mat(labelArr).T
    errCount = 0
    accuracy = 0
    m,n = np.shape(dataMat)
    for iter in range(m):
        xi = dataMat[iter]
        yi = labelMat[iter]
        if -1*yi*(w*xi.T + b) >= 0:#è¯¯åˆ†ç±»
            errCount += 1
    accuracy = 1 - (errCount/m)
    return accuracy
```

```python
import time
def loadData(fileName):
    '''
    åŠ è½½Mnistæ•°æ®é›†
    :param fileName:è¦åŠ è½½çš„æ•°æ®é›†è·¯å¾„
    :return: listå½¢å¼çš„æ•°æ®é›†åŠæ ‡è®°
    '''
    print('start to read data')
    # å­˜æ”¾æ•°æ®åŠæ ‡è®°çš„list
    dataArr = []; labelArr = []
    # æ‰“å¼€æ–‡ä»¶
    fr = open(fileName, 'r')
    # å°†æ–‡ä»¶æŒ‰è¡Œè¯»å–
    for line in fr.readlines():
        # å¯¹æ¯ä¸€è¡Œæ•°æ®æŒ‰åˆ‡å‰²ç¦','è¿›è¡Œåˆ‡å‰²ï¼Œè¿”å›å­—æ®µåˆ—è¡¨
        curLine = line.strip().split(',')

        # Mnsitæœ‰0-9æ˜¯ä¸ªæ ‡è®°ï¼Œç”±äºæ˜¯äºŒåˆ†ç±»ä»»åŠ¡ï¼Œæ‰€ä»¥å°†>=5çš„ä½œä¸º1ï¼Œ<5ä¸º-1
        if int(curLine[0]) >= 5:
            labelArr.append(1)
        else:
            labelArr.append(-1)
        #å­˜æ”¾æ ‡è®°
        #[int(num) for num in curLine[1:]] -> éå†æ¯ä¸€è¡Œä¸­é™¤äº†ä»¥ç¬¬ä¸€å“¥å…ƒç´ ï¼ˆæ ‡è®°ï¼‰å¤–å°†æ‰€æœ‰å…ƒç´ è½¬æ¢æˆintç±»å‹
        #[int(num)/255 for num in curLine[1:]] -> å°†æ‰€æœ‰æ•°æ®é™¤255å½’ä¸€åŒ–(éå¿…é¡»æ­¥éª¤ï¼Œå¯ä»¥ä¸å½’ä¸€åŒ–)
        dataArr.append([int(num)/255 for num in curLine[1:]])

    #è¿”å›dataå’Œlabel
    return dataArr, labelArr
def perceptron(dataArr, labelArr, iter=50):
    dataMat = np.mat(dataArr)
    labelMat = np.mat(labelArr).T
    m,n = np.shape(dataMat)
    w = np.zeros((1,n))
    b = 0
    h = 0.0001
    
    for k in range(iter):
        for i in range(m):
            xi = dataMat[i]#returns vector
            yi = labelMat[i]
            if -1*yi*(w*xi.T+b) >= 0:
                w+=h*yi*xi
                b+=h*yi
        print('Round %d:%d training' % (k, iter))
    return w,b


def model_test(dataArr, labelArr, w, b):
    '''
    æµ‹è¯•å‡†ç¡®ç‡
    :param dataArr:æµ‹è¯•é›†
    :param labelArr: æµ‹è¯•é›†æ ‡ç­¾
    :param w: è®­ç»ƒè·å¾—çš„æƒé‡w
    :param b: è®­ç»ƒè·å¾—çš„åç½®b
    :return: æ­£ç¡®ç‡
    '''
    print('start to test')
    #å°†æ•°æ®é›†è½¬æ¢ä¸ºçŸ©é˜µå½¢å¼æ–¹ä¾¿è¿ç®—
    dataMat = np.mat(dataArr)
    #å°†labelè½¬æ¢ä¸ºçŸ©é˜µå¹¶è½¬ç½®ï¼Œè¯¦ç»†ä¿¡æ¯å‚è€ƒä¸Šæ–‡perceptronä¸­
    #å¯¹äºè¿™éƒ¨åˆ†çš„è§£è¯´
    labelMat = np.mat(labelArr).T

    #è·å–æµ‹è¯•æ•°æ®é›†çŸ©é˜µçš„å¤§å°
    m, n = np.shape(dataMat)
    #é”™è¯¯æ ·æœ¬æ•°è®¡æ•°
    errorCnt = 0
    #éå†æ‰€æœ‰æµ‹è¯•æ ·æœ¬
    for i in range(m):
        #è·å¾—å•ä¸ªæ ·æœ¬å‘é‡
        xi = dataMat[i]
        #è·å¾—è¯¥æ ·æœ¬æ ‡è®°
        yi = labelMat[i]
        #è·å¾—è¿ç®—ç»“æœ
        result = -1 * yi * (w * xi.T + b)
        #å¦‚æœ-yi(w*xi+b)>=0ï¼Œè¯´æ˜è¯¥æ ·æœ¬è¢«è¯¯åˆ†ç±»ï¼Œé”™è¯¯æ ·æœ¬æ•°åŠ ä¸€
        if result >= 0: errorCnt += 1
    #æ­£ç¡®ç‡ = 1 - ï¼ˆæ ·æœ¬åˆ†ç±»é”™è¯¯æ•° / æ ·æœ¬æ€»æ•°ï¼‰
    accruRate = 1 - (errorCnt / m)
    #è¿”å›æ­£ç¡®ç‡
    return accruRate

if __name__ == '__main__':
    #è·å–å½“å‰æ—¶é—´
    #åœ¨æ–‡æœ«åŒæ ·è·å–å½“å‰æ—¶é—´ï¼Œä¸¤æ—¶é—´å·®å³ä¸ºç¨‹åºè¿è¡Œæ—¶é—´
    start = time.time()

    #è·å–è®­ç»ƒé›†åŠæ ‡ç­¾
    trainData, trainLabel = loadData('../Mnist/mnist_train.csv')
    #è·å–æµ‹è¯•é›†åŠæ ‡ç­¾
    testData, testLabel = loadData('../Mnist/mnist_test.csv')

    #è®­ç»ƒè·å¾—æƒé‡
    w, b = perceptron(trainData, trainLabel, iter = 30)
    #è¿›è¡Œæµ‹è¯•ï¼Œè·å¾—æ­£ç¡®ç‡
    accruRate = model_test(testData, testLabel, w, b)

    #è·å–å½“å‰æ—¶é—´ï¼Œä½œä¸ºç»“æŸæ—¶é—´
    end = time.time()
    #æ˜¾ç¤ºæ­£ç¡®ç‡
    print('accuracy rate is:', accruRate)
    #æ˜¾ç¤ºç”¨æ—¶æ—¶é•¿
    print('time span:', end - start)
```
